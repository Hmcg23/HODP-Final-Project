```
import pandas as pd
import matplotlib.pyplot as plt
# Problem 1: Importing such a large data set
# Solution - Add a bunch of rows to a google sheet and then copy and paste from a google doc 
# (working btwn the .docx and the Google Sheet) didn't go too well
df = pd.read_csv("/work/directory.csv")

df.tail()
# checking that I had all the names in the list
# find out what yr this stops at later

df['first_name'] = df['Name'].str.split().str[0]
df['last_name'] = df['Name'].str.split().str[-1]
# splitting into last name and first name

print(df.loc[948,['first_name', 'last_name']])
# checking that the first name and last names come out correct

print(df['Name'].mode())
# this shows us that the most common Harvard name overall is Lauren Park

top20_fn = df['first_name'].value_counts().head(20)

print(top20_fn)

top20_ln = df['last_name'].value_counts().head(20)


top20_fn.plot(kind = 'bar', figsize = (10, 6))
plt.title("Top 20 First Names")
plt.xlabel("First Name")
plt.ylabel("Frequency")
plt.show()
# this shows us the top 20 most common first names at Harvard

top20_ln.plot(kind = 'bar', figsize = (10,6))
plt.title("Top 20 Last Names")
plt.xlabel("Last Name")
plt.ylabel("Frequency")
plt.show()
# this shows us the top 20 most common last names at Harvard... how can we improve on this visualization?




```
